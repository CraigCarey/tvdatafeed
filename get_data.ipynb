```python
from tvDatafeed import TvDatafeed

tv = TvDatafeed()
```

```python
hist = tv.get_hist("GENL", "LSE")
display(hist)
```

```python
symbol = "GENL"
exchange = "LSE"
fut_contract = None

symbol = tv.format_symbol(
    symbol=symbol, exchange=exchange, contract=fut_contract
)
symbol
```

### Get Historic Price Data

```python
extended_session = False
n_bars = 10
interval = "1D"
sentinel = "series_completed"

tv.create_connection()
tv.send_set_auth_token()
tv.send_chart_create_session_msg()
tv.send_quote_create_session_msg()
tv.send_quote_set_fields_overview_msg()
# tv.send_quote_add_symbols_msg(symbol)
tv.send_message(
    "quote_add_symbols",
    [tv.session, symbol],
)
tv.send_resolve_symbol_msg(symbol, extended_session)
tv.send_create_series_msg(interval, n_bars)

raw_data = tv.receive_data(sentinel=sentinel)
```

```python
raw_data
```

```python
tv.create_hist_df(raw_data, symbol)
```

### Get Symbol Data

```python
sentinel = "symbol_resolved"

tv.create_connection()
tv.send_set_auth_token()
tv.send_chart_create_session_msg()
tv.send_resolve_symbol_msg(symbol, extended_session)

raw_data = tv.receive_data(sentinel=sentinel)
raw_data
```

```python
raw_sym_data = tv.parse_symbol_data(raw_data)
raw_sym_data
```
